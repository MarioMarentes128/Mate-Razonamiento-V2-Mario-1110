#### Problema 3
Resta de un númeroEnunciado: Un número al que se le resta siete da como resultado 18. ¿Cuál será ese número?Paso 1: Análisis y ResoluciónIdentificar la incógnita: Sea $x$ el número que buscamos.Plantear la ecuación: "Un número ($x$) al que se le resta ($-$) siete ($7$) da como resultado ($=$) 18".$$x - 7 = 18$$Despejar la variable: El 7 que está restando pasa al otro lado sumando.$$x = 18 + 7$$Resultado:$$x = 25$$

```python
# Problema 3: Encontrar el número original tras una resta
resultado_final = 18
cantidad_restada = 7

# Despejamos: x = resultado_final + cantidad_restada
x = resultado_final + cantidad_restada

print(f"El número buscado es: {x}")
```

####Problema 4
El triple de un númeroEnunciado: El triple de un número más cuatro es igual a 31.Paso 1: Resolución matemáticaVariable: Sea $x$ el número.Ecuación: "El triple de un número ($3x$) más ($+$) cuatro ($4$) es igual ($=$) a $31$".$$3x + 4 = 31$$Despeje (paso 1): El 4 pasa restando:$$3x = 31 - 4$$$$3x = 27$$Despeje (paso 2): El 3 que multiplica a la $x$ pasa dividiendo:$$x = \frac{27}{3}$$Resultado: $x = 9$

```python
# Problema 4: Ecuación con multiplicador y suma
total = 31
termino_independiente = 4
multiplicador = 3

# Paso 1: Restar el 4 al total
# Paso 2: Dividir entre el multiplicador (3)
x = (total - termino_independiente) / multiplicador

print(f"El número buscado es: {int(x)}")
```

#### Problema 5
Dimensiones de un terreno rectangularEnunciado: El largo mide 5m más que el ancho. Perímetro total = 50m.Paso 1: Resolución matemáticaVariables:Ancho: $x$Largo: $x + 5$Ecuación del perímetro: El perímetro es la suma de los 4 lados ($2 \cdot \text{largo} + 2 \cdot \text{ancho}$):$$2(x + 5) + 2(x) = 50$$Simplificación:$$2x + 10 + 2x = 50$$$$4x + 10 = 50$$Despeje:$$4x = 50 - 10$$$$4x = 40$$$$x = \frac{40}{4} = 10$$Resultados:Ancho: 10 metros.Largo: $10 + 5 =$ 15 metros.

```python
# Problema 5: Dimensiones de un terreno rectangular
perimetro = 50
diferencia_largo = 5

# La ecuación simplificada es 4x + (2 * diferencia) = perimetro
# Despejamos x (ancho):
ancho = (perimetro - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"El ancho mide: {ancho}m")
print(f"El largo mide: {largo}m")
```
